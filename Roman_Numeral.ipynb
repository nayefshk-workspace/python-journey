```python
#Roman numeral dictionary.
roman = {"I":1,
         "V":5,
         "X":10,
         "L":50,
         "C":100,
         "D":500,
         "M":1000}
```

```python
#total var to add at the end
total = 0
```

```python
#checking dictionary
print(roman["I"])   # 1
print(roman["M"])   # 5
```

```python
#input
roman_number = str(input("Enter your roman number: "))
```

```python
#creating a loop for index
for i in range(len(roman_number)):
  roman_number[i]
```

```python
#making the current variable
current = roman_number[i]
type(current)
current = roman[roman_number[i]] #then converting it in into a integer
type(current)
```

```python
# Adding conditions
# If the current value is smaller than the next value, subtract it.
# Otherwise add it
if i < len(roman_number) - 1:
  next_value = roman[roman_number[i + 1]]
  current < next_value
  total -= current
else:
  total = total + current
```

```python
# printing answer
print(total)
```
