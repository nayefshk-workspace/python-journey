```python
# Write a program in Python to find the greatest among three integers. 
# ● Example: 
# ○ Input: 10, 20, 30 
# ○ Output: 30 
# 
first_number = int(input("Enter your first number: "))
second_number = int(input("Enter your second number: "))
third_number = int(input("Enter your third number: "))


if first_number >= second_number and first_number >= third_number:
    greatest = first_number
elif second_number >= first_number and second_number >= third_number:
    greatest = second_number
else:
    greatest = third_number

print(f"The greatest number is {greatest}")
```

```python
# Write a program in Python to check if a given number is prime or not. 
# ● Example: 
# ○ Input: 17 
# ○ Output: True
num = int(input("Enter your number"))

is_prime = True
for  i in range(2,(num//2)+1):
    if num % i == 0:
        is_prime = False
        break
    else:
        continue
if is_prime:
    print(f"{num} is prime ")
else:
    print(f"{num} is not prime ")
```

```python
# Write a program to reverse an integer in Python. 
# ● Example: 
# ○ Input: 1234 
# ○ Output: 4321 

num = int(input("Enter a number: "))
n = num
reverse = 0 

while n != 0:
    digit = n % 10
    reverse = reverse * 10  + digit
    n = n // 10
print(reverse)



```
